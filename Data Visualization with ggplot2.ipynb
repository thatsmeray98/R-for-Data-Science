# Aesthetic Mappings

```r
ggplot(data = mpg) + 
 geom_point(mapping = aes(x = displ, y = hwy, color =  class))
```

This color-coded scatterplot of the variables engine size (displ) and fuel efficiency (hwy), uses the "class" categorical variable to display the respective class of each car.

```r
ggplot(data = mpg) + 
 geom_point(mapping = aes(x = displ, y = hwy, size = class))
```

Its not a good idea to map an unordered categorical variable ("class") to an ordered asthetic ("size").

```r
ggplot(mpg) + 
 geom_point(mapping = aes(x = displ, y = hwy, alpha = class))
```

```r
ggplot(mpg, aes(x = displ, y = hwy, shape = class)) + 
 geom_point()
```

In the above plot the "suv" group is not plotted as, ggplot only uses 6 shapes at a time. By default, all the other groups will remain unplotted in the graph when shape asthetic is used.

```r
ggplot(mpg) + 
 geom_point(mapping = aes(x = displ, y = hwy), color = 'blue')
```

This graph shows that, the asthetic property of the geom function can also be set manually. Set the asthetic manually by name as an arguement of the geom function; i.e. it goes outside of aes().

```r
ggplot(mpg , aes(displ, cty, color = cty < 17, size = displ)) + 
 geom_point()
```

Aesthetics can also be mapped to expressions like cty < 17. The ggplot() function behaves as if a temporary variable was added to the data with values equal to the result of the expression. In this case, the result of cty < 17 is a logical variable which takes values of TRUE or FALSE.

# Facets

A way to add additional variables and its perticularly useful for categorical variables. It splits the plot into facets, subplots that each display one subset of the data.

```r
ggplot(mpg , aes(displ, hwy)) + 
 geom_point() + 
 facet_wrap(~class, nrow = 3)
```

```r
ggplot(mpg , aes(displ, hwy)) + 
 geom_point() + 
 facet_wrap(~manufacturer , nrow = 3)
```

```r
ggplot(mpg , aes(displ, cty)) + 
 geom_point() + 
 facet_grid(drv~cyl)
```

```r
ggplot(mpg, aes(displ, hwy)) + 
 geom_point() + 
 facet_grid(class~fl)
```

```r
ggplot(mpg, aes(displ, cty)) + 
 geom_point() + 
 facet_grid(.~drv)
```

```r
ggplot(mpg, aes(displ, hwy)) + 
 geom_point() + 
 facet_wrap(~cty , nrow = 3)
```

If continuous variable is used in a facet then, its converted into a categorical variable and each unique value is plotted as a different group.

# Position Adjustments

```r
ggplot(diamonds, aes(cut, fill = cut)) + 
 geom_bar()
```

```r
ggplot(diamonds, aes(cut, color = cut)) + 
 geom_bar()
```

color aesthetic colors the boundaries of the bars that are plotted with the help of geom_bar function whereas, the fill aesthetic fills the whole bar area that are plotted with the help of geom_bar function.

```r
ggplot(diamonds, aes(cut, fill = clarity)) +
 geom_bar()
```

If we map fill aesthetic with another categorical variable then, the bars get automatically stacked where each colored rectangle in the bars represent a combination of the two variables (like, in the above graph each rectangle in the bars represent a combination of the variables cut and clarity).

This stacking is performed automatically by postion adjustment specified by the position argument. Setting the position manually to identity will place each object exactlly where it falls on the graph. This is not a good idea for barplots as, it makes them overlap within each bar. So to overcome, either make the bars slightly transparent by using small alpha value or completely transparent by setting fill = NA (use color aesthetic to show the boundaries of the rectangles in each of the bars).

```r
ggplot(diamonds, aes(cut, fill = clarity)) + 
 geom_bar(alpha = 1/5, position = "identity")
```

```r
ggplot(diamonds, aes(cut, color = clarity)) + 
 geom_bar(fill = NA, position = 'identity')
```

The identity position is more useful in 2D geoms like geom_point, where its the default.

Next, the position can also be set to 'fill', where each set of stacked bar is of same height (This is a standardized stacked barplot). This is useful where need is to compare proportions across groups.

```r
ggplot(diamonds, aes(cut, fill = clarity)) +
 geom_bar(position = 'fill')
```

Lastly, the position can also be 'dodge', where it plots overlapping objects directly beside each other. This makes it easier to compare individual values. This is a side-by-side barplot.

```r
ggplot(diamonds, aes(cut, fill = clarity)) + 
 geom_bar(position = 'dodge')
```

NOTE :- There one more adjustment to position which is very useful for scatterplots. The values in scatterplots are rounded so that the points could appear on the grid and don't overlap each other (this problem is called as overplotting). This arrangement makes it harder to see "where the mass of the data lies?". To avoid this gridding, we can set the position adjustment to 'jitter'. The position = 'jitter' adds a small amount of random noise to each point.

```r
ggplot(mpg, aes(displ, hwy)) + 
 geom_point(position = 'jitter') +
 geom_point(color = 'red') +
 geom_jitter(height = 0.25, width = 0.25, color = 'green')
```

Adding randomness makes the graph less accurate if working at small scale but, it also makes the graph more revealing if working at large scales.

Note :- The height and width arguments are in the units of the data. Thus height = 1 (width = 1) corresponds to different relative amounts of jittering depending on the scale of the y (x) variable. The default values of height and width are defined to be 80% of the resolution() of the data, which is the smallest non-zero distance between adjacent values of a variable. When x and y are discrete variables, their resolutions are both equal to 1, and height = 0.4 and width = 0.4 since the jitter moves points in both positive and negative directions.

# Coordinate Systems

Default coordiate system is teh cartesian coordinate system(x and y act independently to find teh position of a point). 

Other coordinate systems include : coord_flip(), coord_quickmap(), coord_polar()

coord_flip() : It flips the x and y axes. Useful for horizontal boxplots and plotting long labels.

```r
ggplot(mpg, aes(class, hwy)) + 
 geom_boxplot() +
 coord_flip()
```

coord_quickmap() : It sets the aspect ratio accordingly for maps.

```r
nz = map_data('nz')

ggplot(nz, aes(long, lat, group = group)) +
 geom_polygon(fill = 'white', color = 'black') +
 coord_quickmap()
```

coord_polar() : This uses polar coordinates. Polar coordinates reveal an intresting connection between the barplot and the coxcomb plot

```r
bar = (
 ggplot(diamonds, aes(cut, fill = cut)) +
  geom_bar(
      show.legend = FALSE,
      width = 1
  )
)

bar + coord_flip()

bar + coord_polar()
```

```r
ggplot(mpg, aes(factor(1), fill = drv)) + 
 geom_bar(
     width = 1
 ) +
 coord_polar(theta = 'y')
```

```r
ggplot(mpg, aes(cty, hwy)) + 
 geom_point() + 
 geom_abline() + 
 coord_fixed()
```

The function coord_fixed() ensures that the line produced by geom_abline() is at a 45-degree angle. A 45-degree line makes it easy to compare the highway and city mileage to the case in which city and highway MPG were equal. If we didn’t include geom_coord(), then the line would no longer have an angle of 45 degrees.
